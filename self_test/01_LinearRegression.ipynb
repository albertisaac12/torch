```python
# Create randomdata
# train_test split the data
# train a linearRegression Model
# predit values
```

```python
X = torch.arange(1,101,0.5).unsqueeze(dim=1)
X[:5],X.shape
```

```python
torch.manual_seed(42)
weight = torch.randn(1,dtype=torch.float)
bias = torch.randn(1,dtype=torch.float)

y = weight*X + bias
weight,bias
```

```python
train_size = int(len(X) * 0.8)
train_size
```

```python
X_train,y_train = X[:train_size],y[:train_size]
X_test,y_test = X[train_size:],y[train_size:]

X_train.size(), X_test.size(), y[:5]
```

```python
class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        torch.manual_seed(50)
        self.weight = nn.Parameter(torch.randn(1,dtype=torch.float,requires_grad=True))
        self.bias = nn.Parameter(torch.randn(1,dtype=torch.float))

    def forward(self,x:torch.Tensor) -> torch.Tensor:
        return self.weight*x + self.bias
        
```

```python
model0 = LinearRegression()
```

```python
list(model0.parameters())
```

```python
model0.state_dict()
```

```python
with torch.inference_mode():
    y_pred = model0(X_test)

loss = y_test[:5] - y_pred[:5]
y_pred[:5],y_test[:5],loss
```

```python
# inference mode turns of gradient tracking
```

```python
# Model Training  => loss and optimizer
from torch.nn import L1Loss
from torch.optim import Adam

loss_fn = L1Loss()
opt_fn = Adam(params=model0.parameters(),lr=0.01)
```

```python
# Very important Flow
# Forward Pass
# Claculate Loss
# Zero Gradients
# Perfrom Backpropagation
# Update the optimizer
```

```python
epochs = range(0,101)

for epoch in epochs:
    
    model0.train()
    y_pred = model0(X_train)
    loss = loss_fn(y_pred,y_train)
    opt_fn.zero_grad()
    loss.backward()
    opt_fn.step()
```

```python
model0.eval()

with torch.inference_mode():
    y_pred = model0(X_test)
```

# Stacking Notes

`You index into the stack by entering the dimesion for instance when dim =2 you enter the secent dimension and then choose an element each from the dimesion and then stack it` 

```python
# torch.cat()
# torch.split()
# torch.chunk()

torch.manual_seed(42)
x = torch.randn(2,3)
torch.manual_seed(43)
y = torch.randn(2,3)
torch.manual_seed(44)
z = torch.randn(2,3)

x,y,z
```

```python
torch.cat([x,y,z],dim=1)
```

```python
help(torch.chunk)
```

```python
x

s = torch.stack([x,x,x],dim=0)
s, x.shape , s.shape
```

```python
# torch.chunk() # will result in a view of the original tensor while
torch.chunk(s, chunks = 3, dim = 2)
```

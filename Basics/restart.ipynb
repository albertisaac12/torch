```python
def gen_demo():
    yield "first statement"
    yield "seccond statement"
    return "third statement"

# 1. Create the generator object
g = gen_demo()

# 2. Get first value
print(next(g))  # Output: "first statement"

# 3. Get second value
print(next(g))  # Output: "seccond statement"

# 4. Reach the return statement
try:
    print(next(g))
except StopIteration as e:
    print(f"Generator ended. Return value: {e.value}")
    # Output: "Generator ended. Return value: third statement"
```

# Python Multithreading

What is a thread ?

what is multi threading ?

what is multi processing ?


```python
# classes and fucntions can be used to make threads
```

```python
from threading import Thread

class Hello:
    def do(self):
        for i in range(5):
            print("Hello",i+1)

class Hi:
    def do(self):
        for i in range(5):
            print("Hi",i+1)
```

```python
t1 = Hello()
t2 = Hi()
```

```python
t1.do()
```

```python
t2.do()
```

```python
# i want to run hello and hi together

t1 = Thread(target=Hello.do)
t2 = Thread(target=Hi.do)
```

```python
class Hello(Thread):
    def run(self):
        for i in range(5):
            print("Hello",i+1)

class Hi(Thread):
    def run(self):
        for i in range(5):
            print("Hi",i+1)
```

```python
t1 = Hello()
t2 = Hi()
```

```python
t1.start()
t2.start()

print("bye")
```
